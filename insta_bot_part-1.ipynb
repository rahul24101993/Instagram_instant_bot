# Q1 
# Login to your Instagram Handle

# Q1.1
# Submit with sample username and password


```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
driver = webdriver.Chrome(executable_path = 'C:/python/chromedriver')
driver.get('https://www.instagram.com/')
def login():             #function to login into instagram
    wait = WebDriverWait(driver, 10)
    username = wait.until(EC.presence_of_element_located((By.NAME, 'username'))) #waiting unless the username box is avaialble
    username.send_keys("Sample Username")         #send the username
    password=driver.find_element_by_name("password")      #box for password
    password.send_keys("Sample Password")          #sending password
    login=driver.find_element_by_xpath("//*[@id='loginForm']/div/div[3]/button/div")  #login button
    login.submit()                         #click on login button
login()
```

# Q5
# Like/Unlike posts

# Q5.1
# Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.


```python
def Like():
    import time
    from bs4 import BeautifulSoup
    driver.get('https://www.instagram.com/')
    time.sleep(4)
    driver.find_element_by_xpath("//body").click()
    search=driver.find_element_by_class_name("XTCLo")
    search.clear()
    search.send_keys("dilsefoodie")
    time.sleep(3)
    drp_dwn = driver.find_element_by_class_name('Ap253')
    drp_dwn.click()
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, 2000)") 
    time.sleep(2)
    base='https://www.instagram.com'
    from bs4 import BeautifulSoup
    posts=[]
    e=driver.find_elements_by_class_name('v1Nh3')           #class for each posts
    count=0
    for j in e:
        if(count==30):             #count to have 30 posts
            break
        source=j.get_attribute('innerHTML')              #getting the urls of each posts
        data=BeautifulSoup(source, 'html')               #parsing using beautifulsoup
        posts.append(base+data.a['href'])
        count+=1
    var=1
    for i in posts:
        driver.get(i)
        m=driver.find_elements_by_class_name('_8-yf5')
        source=m[1].get_attribute('outerHTML')
        data=BeautifulSoup(source, 'html')
        if(data.html.body.svg['aria-label']=='Unlike'):          #check whether the post is liked or not
            print(var,'----> post already liked')
        else:
            m=driver.find_elements_by_class_name('QBdPU')  # classes to get the like button
            m[1].click()                #clicking to like the posts
        var+=1
Like()    
```

```python
def followers():
    common_followers=[]                            #list to fill the common follwers of the user and foodtalkindia
    import time
    from bs4 import BeautifulSoup
    driver.get('https://www.instagram.com/')
    time.sleep(4)
    driver.find_element_by_xpath("//body").click()
    search=driver.find_element_by_class_name("XTCLo")
    search.clear()
    search.send_keys("foodtalkindia")                          #searching for 'foodtalkindia'
    time.sleep(2)
    drp_dwn = driver.find_element_by_class_name('Ap253')      #choosing the first one from the dropdown
    drp_dwn.click()
    time.sleep(4)
    e=driver.find_elements_by_class_name('g47SY')
    e[1].click()
    time.sleep(2)
    count=0
    while True:
        driver.find_element_by_class_name("isgrP").click()
        driver.find_element_by_tag_name("body").send_keys(Keys.END)       #scrolling the followers window
        e=driver.find_elements_by_class_name('FPmhX')            #class having details of the username of followers
        x=driver.find_elements_by_class_name('sqdOP')             #class having details wheather the user is following or not 
        if(len(e)>500):              #to have min first 500 followers
            break
        if(count>200):
            break
        count+=1
    count=0
    while(count<501):
        if(x[count].get_attribute('innerHTML')=='Following'):         #having a check weather the user is following that follower or not
            common_followers.append(e[count].get_attribute('innerHTML'))  #if following than append it to the list common_follower
        count+=1
    for i in common_followers:
        print(i)                             #finally printing the common_followers
followers()                 
```

```python
def check_cn_story():
    driver.get("https://www.instagram.com/")
    time.sleep(3)
     
    list_stories=(driver.find_elements_by_xpath('//*[@id="react-root"]/section/main/section/div[1]/div[1]/div/div/div/div/ul/li'))[2:]
    list_canvas=driver.find_elements_by_xpath('//*[@id="react-root"]/section/main/section/div[1]/div[1]/div/div/div/div/ul/li//canvas')
    
    count=0
    for story in list_stories:
        
        if story.text=="coding.ninjas":
            if list_canvas[count].get_attribute("height")=="66":
                print("Visit Web driver Session, Story has been opened there!")
                story.click()
            else:
                print("Story has been Seen Already!")
            break
        count+=1
    else:
        print("Story Not Available!")

check_cn_story()
```
